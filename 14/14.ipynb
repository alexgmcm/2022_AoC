```python
def str2tuple(string):
    string=string.split(",")
    return (int(string[0]),int(string[1]))

def line2coords(line):
    line = line.split("->")
    return [str2tuple(x.strip()) for x in line]
```

```python
#visualise cave
from functools import reduce
def visualise_cave(grid):
    cave=[]
    x_min = min([a[0] for a in grid.keys()])
    x_max = max([a[0] for a in grid.keys()])
    y_min = 0 #min([a[1] for a in grid.keys()])
    y_max = max([a[1] for a in grid.keys()])

    x_range=range(x_min, x_max+1)
    y_range=range(y_min,y_max+1)

    for y in y_range:
        cave.append([])
        for x in x_range:
            if (x,y) in grid:
                char = grid[(x,y)]
            elif (x,y) == (500,0):
                char = "+"
            else:
                char = "."
            cave[y].append(char)
    return cave
```

```python
from math import copysign
TEST_FILE="1_test.txt"
INPUT_FILE="1_input.txt"
file=TEST_FILE
f=open(file,'r')
lines=f.readlines()
coords_list=[]
grid={}

for line in lines:
    line=line.strip()
    coords=line2coords(line)
    coords_list.append(coords)
    
for coords in coords_list:
    for i,coord in enumerate(coords):
        if i!=len(coords)-1:
            # not last coord
            next_coord=coords[i+1]
            #print(int(copysign(1, next_coord[0]-coord[0])))
            x_direction = int(copysign(1, next_coord[0]-coord[0]))
            y_direction = int(copysign(1, next_coord[1]-coord[1]))
            x_range = range(coord[0],next_coord[0] + x_direction, x_direction) 
            y_range = range(coord[1],next_coord[1] + y_direction, y_direction)
            #print(x_range)
            #print(y_range)
            coords_to_fill = [(x,y) for x in x_range for y in y_range ]
            #print(coords_to_fill)
            for target in coords_to_fill:
                grid[target]="#" 
   
```

## Part 1

```python
y_max = max([a[1] for a in grid.keys()])
sand_queue = []
settled_ct = 0
finish=False

while True:
    if len(sand_queue) == 0:
        sand_queue.append((500,0))
    sand=sand_queue.pop()
    if file == TEST_FILE:
        if settled_ct in [1,2,5,22,24]:
            print(f"Sand Count: {settled_ct}")
            for row in visualise_cave(grid):
                print(row)
            print("----------")
    while True:
        moved=False
        down_move = (sand[0],sand[1]+1)
        diagonal_left = (sand[0]-1,sand[1]+1)
        diagonal_right = (sand[0]+1,sand[1]+1)
        moves = [down_move, diagonal_left, diagonal_right]
        for move in moves:
            #print(f"Trying move from {sand} to {move}")
            if move not in grid:
                grid[move] = "." # not visited before, must be air
            if grid[move] == ".":
                sand = move
                moved = True
                break
        if not moved:
            grid[sand] = "o" #settled sand
            settled_ct += 1
            break
        if sand[1] > y_max:
            finish=True
            break #sand is now falling to infinity?
    if finish:
        break
print("----------")
print(f"Units of sand at rest: {settled_ct}")
print("----------")
```
